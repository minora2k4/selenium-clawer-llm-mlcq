````python
import pandas as pd
from pprint import pprint
import json
from typing import List, Dict, Any
````

````python
df = pd.read_json("../output/questions.json")
data = df.to_dict()
process_data = data.copy()
````

````python
# Build prompt
def build_prompt(dict : dict) -> str:
    prompt = f"""
    You are an expert translator specializing in Machine Learning and Artificial Intelligence.

    Translate the following Python dictionary from English to Vietnamese.

    Requirements:
    1. Return a new Python dictionary with exactly the same structure, keys, and nesting.
    2. Translate only the string values containing natural language.
    3. Do NOT translate dictionary keys, option labels (`a`, `b`, `c`, `d`), variable names, mathematical expressions, LaTeX, Markdown, placeholders, or code.
    4. Keep core Machine Learning and Deep Learning technical terms in their original English form if translating them makes them sound unnatural or lose their precise technical definition (e.g., KEEP 'Gradient Descent', 'Overfitting', 'Underfitting', 'Backpropagation', 'Epoch', 'Batch', 'Transformer', 'Embedding'). Do NOT translate 'Gradient Descent' into 'giảm gradient'.
    5. Preserve all mathematical expressions, variables, symbols, Markdown, LaTeX, escaped characters, and formatting exactly as they appear.
    6. Use standard Vietnamese terminology for general context but strict technical terms for ML/AI components.
    7. Do not rewrite, simplify, or expand the original content. Produce a faithful translation only.
    8. Return ONLY the raw Python dictionary content. Do NOT wrap the output in Markdown code blocks (like ```python) and do not include any conversational text or explanations.

    With this Example:
    I have Input:
    {{
        'answer': 'c',
        'explanation': '$t_{{0}}$ and $t_{{1}}$ try to minimize prediction error on the training set. Since y is the target variable, h(x) must be y (ideally) or closer to y.',
        'options': {{
            'a': 'Give negative h(x)',
            'b': 'Give h(x) as close to 0 as possible, without themselves being 0',
            'c': 'Give h(x) as close to y, in training data, as possible',
            'd': 'Give h(x) closer to x than y'
        }},
        'question': 'The hypothesis is given by h(x) = $t_{{0}}$ + $t_{{1}}$x. What is the goal of $t_{{0}}$ and $t_{{1}}$?'
    }}
    With this Input so Output is:
    {{
        'answer': 'c',
        'explanation': 'Các tham số $t_{{0}}$ và $t_{{1}}$ cố gắng giảm thiểu sai số dự đoán trên tập dữ liệu huấn luyện. Vì y là biến mục tiêu, nên h(x) lý tưởng sẽ bằng y, hoặc ít nhất càng gần y càng tốt.',
        'options': {{
            'a': 'Cho ra giá trị h(x) âm',
            'b': 'Cho ra giá trị h(x) gần 0 nhất có thể, nhưng bản thân chúng không bằng 0',
            'c': 'Cho ra giá trị h(x) gần với y trong dữ liệu huấn luyện nhất có thể',
            'd': 'Cho ra giá trị h(x) gần với x hơn là y'
        }},
        'question': 'Giả sử h(x) = $t_{{0}}$ + $t_{{1}}$x. Mục tiêu của $t_{{0}}$ và $t_{{1}}$ là gì?'
    }}

    Task:
    Translate the input {dict} dictionary below into Vietnamese following the exact same rules and structure. Output ONLY the valid dictionary.

    Output:
    """
    return prompt
````

````python
# Call LLM
import ast
import time
import requests
from dotenv import load_dotenv
import os

load_dotenv()  # Đọc file .env

API_URL = "https://openrouter.ai/api/v1/chat/completions"
MODEL = "tencent/hy3:free"
api_key = os.getenv("YOUR_API_KEY")
HEADERS = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json",
}

def call_llm(input: dict):
    prompt = build_prompt(input)

    while True:
        try:
            print(f"Trying model: {MODEL}")
            resp = requests.post(
                API_URL,
                headers=HEADERS,
                json={
                    "model": MODEL,
                    "messages": [
                        {
                            "role": "user",
                            "content": prompt,
                        }
                    ],
                    "reasoning": {"enabled": False},
                },
                timeout=60,
            )
            resp.raise_for_status()
            data = resp.json()
            result = ast.literal_eval(
                data["choices"][0]["message"]["content"]
            )
            print("Success")
            return result
        except Exception as e:
            print(f"Error: {e}")
            print("Retry after 10s...")
            time.sleep(10)
````

````python
# Run call_llm
for k, v in process_data.items():
    if k == "html":
        for i in range(len(process_data[k])):
            for j in range(len(process_data[k][i])):
                process_data[k][i][j] = call_llm(process_data[k][i][j])
                print(f"Xử lí title thứ: {i+1}/46")
                print(f"Xử lí xong {j+1} / {len(process_data[k][i])}")
````
