```python
import re
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By


QUESTION_PATTERN = re.compile(r"^\d+\.")
OPTION_PATTERN = re.compile(r"^[a-d]\)")
FIRST_OPTION_PATTERN = re.compile(r"\ba\)\s")
OPTION_MARKER_PATTERN = re.compile(r"\b([a-d])\)\s")

CLOUDFLARE_INDICATORS = [
    "just a moment",
    "checking your browser",
    "cf-browser-verification",
    "cf_chl_",
    "attention required",
    "verify you are human",
]


def has_class_contains(tag, keyword):
    """Kiểm tra class của tag có chứa keyword."""
    return any(keyword in cls for cls in (tag.get("class") or []))


def extract_answer_from_bold(tag):
    """Lấy đáp án từ thẻ <b>/<strong>."""
    bold = tag.find(["b", "strong"])
    if not bold:
        return ""

    text = bold.get_text(" ", strip=True)
    match = re.match(r"^([a-dA-D])\)", text)
    return match.group(1).lower() if match else text


def parse_question(text):
    """Tách câu hỏi và các lựa chọn."""
    text = text.replace("View Answer", "").strip()

    match = FIRST_OPTION_PATTERN.search(text)
    if not match:
        return text, []

    question = text[:match.start()].strip()
    options_text = text[match.start():]

    markers = list(OPTION_MARKER_PATTERN.finditer(options_text))
    options = []

    for i, marker in enumerate(markers):
        start = marker.start()
        end = markers[i + 1].start() if i + 1 < len(markers) else len(options_text)
        options.append(options_text[start:end].strip())

    return question, options


def is_cloudflare_challenge():
    """Kiểm tra xem trang hiện tại có đang hiển thị Cloudflare challenge không."""
    try:
        title = (driver.title or "").lower()
        page_source = driver.page_source.lower()
    except Exception:
        return False

    return any(
        indicator in title or indicator in page_source
        for indicator in CLOUDFLARE_INDICATORS
    )


def wait_for_cloudflare(auto_timeout=15, manual=True):
    """
    Chờ Cloudflare challenge biến mất.
    - auto_timeout: số giây chờ tự động (challenge JS tự giải) trước khi hỏi người dùng.
    - manual: nếu True, sẽ dừng script và chờ bạn giải tay trên cửa sổ trình duyệt
      khi tự động chờ không ăn thua.
    """
    if not is_cloudflare_challenge():
        return

    print("-> Phát hiện Cloudflare challenge, đang chờ...")
    start = time.time()

    while is_cloudflare_challenge():
        if time.time() - start > auto_timeout:
            if manual:
                input("   Vui lòng giải Cloudflare thủ công trên cửa sổ trình duyệt, "
                      "sau đó nhấn Enter tại đây để tiếp tục...")
            break
        time.sleep(1)

    if is_cloudflare_challenge():
        print("-> Vẫn còn challenge, có thể cần thử lại.")
    else:
        print("-> Đã qua Cloudflare, tiếp tục.")


def expand_all_answers():
    """Mở toàn bộ phần đáp án trên trang."""
    buttons = driver.find_elements(
        By.XPATH,
        "//*[contains(@class,'collapseomatic') and not(contains(@class,'collapseomatic_content'))]"
    )

    if not buttons:
        buttons = driver.find_elements(
            By.XPATH,
            "//*[contains(text(),'View Answer')]"
        )

    print(f"-> Mở {len(buttons)} đáp án...")

    for button in buttons:
        try:
            driver.execute_script(
                "arguments[0].scrollIntoView({block:'center'});",
                button
            )
            time.sleep(0.05)
            driver.execute_script("arguments[0].click();", button)
        except Exception:
            pass

    time.sleep(1.5)


def parse_answer_block(block, question):
    """Cập nhật đáp án và giải thích từ khối đáp án."""
    lines = [
        line.strip()
        for line in block.get_text("\n").split("\n")
        if line.strip()
    ]

    for line in lines:
        lower = line.lower()

        if lower.startswith("answer:"):
            question["answer"] = line.split(":", 1)[1].strip()

        elif lower.startswith("explanation:"):
            question["explanation"] = line.split(":", 1)[1].strip()

    if not question["answer"]:
        question["answer"] = extract_answer_from_bold(block)
```

```python
from bs4 import BeautifulSoup

def crawl_html(url, debug=False, force_reload=False):
    """Crawl nguyên HTML block của từng câu hỏi, parse sau."""

    if force_reload or driver.current_url.rstrip("/") != url.rstrip("/"):
        driver.get(url)
        time.sleep(2)
        wait_for_cloudflare()

    expand_all_answers()

    soup = BeautifulSoup(driver.page_source, "lxml")
    content = soup.find("div", class_="entry-content")

    if not content:
        return []

    elements = content.find_all(["p", "div"], recursive=False)

    html_blocks = []

    current_block = []

    for element in elements:

        text = element.get_text(" ", strip=True)

        # bắt đầu câu hỏi mới
        if (
            element.name == "p"
            and QUESTION_PATTERN.match(text)
        ):

            # lưu block trước
            if current_block:
                html_blocks.append(current_block)

            current_block = [element]

        elif current_block:

            # nếu đang trong câu hỏi thì gom tất cả
            current_block.append(element)


    # block cuối
    if current_block:
        html_blocks.append(current_block)

    return html_blocks
```

```python
def render_markdown(html_blocks):
    md = []

    for i, item in enumerate(html_blocks, 1):

        md.append(f"# Question {i}\n")

        md.append("## Question")
        md.append(html_to_markdown(item["question_html"]))

        if item["option_html"]:
            md.append("\n## Options")
            for option in item["option_html"]:
                md.append(html_to_markdown(option))

        if item["answer_html"]:
            md.append("\n## Answer")
            md.append(html_to_markdown(item["answer_html"]))

        md.append("\n---\n")

    return "\n".join(md)
```
