```python
import re
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By


QUESTION_PATTERN = re.compile(r"^\d+\.")
OPTION_PATTERN = re.compile(r"^[a-d]\)")
FIRST_OPTION_PATTERN = re.compile(r"\ba\)\s")
OPTION_MARKER_PATTERN = re.compile(r"\b([a-d])\)\s")

CLOUDFLARE_INDICATORS = [
    "just a moment",
    "checking your browser",
    "cf-browser-verification",
    "cf_chl_",
    "attention required",
    "verify you are human",
]

def extract_answer_from_bold(tag):
    """Lấy đáp án từ thẻ <b>/<strong>."""
    bold = tag.find(["b", "strong"])
    if not bold:
        return ""

    text = bold.get_text(" ", strip=True)
    match = re.match(r"^([a-dA-D])\)", text)
    return match.group(1).lower() if match else text

# Check cloudflare
def is_cloudflare_challenge():
    """Kiểm tra xem trang hiện tại có đang hiển thị Cloudflare challenge không."""
    try:
        title = (driver.title or "").lower()
        page_source = driver.page_source.lower()
    except Exception:
        return False

    return any(
        indicator in title or indicator in page_source
        for indicator in CLOUDFLARE_INDICATORS
    )


def wait_for_cloudflare(auto_timeout=15, manual=True):
    """
    Chờ Cloudflare challenge biến mất.
    - auto_timeout: số giây chờ tự động (challenge JS tự giải) trước khi hỏi người dùng.
    - manual: nếu True, sẽ dừng script và chờ bạn giải tay trên cửa sổ trình duyệt
      khi tự động chờ không ăn thua.
    """
    if not is_cloudflare_challenge():
        return

    print("-> Phát hiện Cloudflare challenge, đang chờ...")
    start = time.time()

    while is_cloudflare_challenge():
        if time.time() - start > auto_timeout:
            if manual:
                input("   Vui lòng giải Cloudflare thủ công trên cửa sổ trình duyệt, "
                      "sau đó nhấn Enter tại đây để tiếp tục...")
            break
        time.sleep(1)

    if is_cloudflare_challenge():
        print("-> Vẫn còn challenge, có thể cần thử lại.")
    else:
        print("-> Đã qua Cloudflare, tiếp tục.")

#Click "show answer" button
def expand_all_answers():
    """Mở toàn bộ phần đáp án trên trang."""
    buttons = driver.find_elements(
        By.XPATH,
        "//*[contains(@class,'collapseomatic') and not(contains(@class,'collapseomatic_content'))]"
    )

    if not buttons:
        buttons = driver.find_elements(
            By.XPATH,
            "//*[contains(text(),'View Answer')]"
        )

    print(f"-> Mở {len(buttons)} đáp án...")

    for button in buttons:
        try:
            driver.execute_script(
                "arguments[0].scrollIntoView({block:'center'});",
                button
            )
            time.sleep(0.05)
            driver.execute_script("arguments[0].click();", button)
        except Exception:
            pass

    time.sleep(1.5)

#Get answer and explaination after click "show answer" button
def parse_answer_block(block, question):
    """Cập nhật đáp án và giải thích từ khối đáp án."""
    lines = [
        line.strip()
        for line in block.get_text("\n").split("\n")
        if line.strip()
    ]

    for line in lines:
        lower = line.lower()

        if lower.startswith("answer:"):
            question["answer"] = line.split(":", 1)[1].strip()

        elif lower.startswith("explanation:"):
            question["explanation"] = line.split(":", 1)[1].strip()

    if not question["answer"]:
        question["answer"] = extract_answer_from_bold(block)
```

```python
#Crawling
def crawl_html(url, debug=False, force_reload=False):
    """Crawl nguyên HTML block của từng câu hỏi, parse sau."""
    if force_reload or driver.current_url.rstrip("/") != url.rstrip("/"):
        driver.get(url)
        time.sleep(2)
        wait_for_cloudflare()
    expand_all_answers()
    soup = BeautifulSoup(driver.page_source, "lxml")
    content = soup.find("div", class_="entry-content")
    if not content:
        return []
    elements = content.find_all(["p", "div"], recursive=False)
    html_blocks = []
    current_block = []
    for element in elements:
        text = element.get_text(" ", strip=True)
        # bắt đầu câu hỏi mới
        if (
            element.name == "p"
            and QUESTION_PATTERN.match(text)
        ):
            # lưu block trước
            if current_block:
                html_blocks.append(current_block)
            current_block = [element]
        elif current_block:
            # nếu đang trong câu hỏi thì gom tất cả
            current_block.append(element)
    # block cuối
    if current_block:
        html_blocks.append(current_block)
    return html_blocks
```

```python
#Crawl data
results = []
for i in range(len(urls)):
    html_blocks = crawl_html(urls[i]["url"])
    result = {"title" : urls[i]["title"],
              "html": html_blocks}
    results.append(result)
```

```python
import re
from bs4 import BeautifulSoup, Tag

def clean_soup(nodes):
    wrapper_soup = BeautifulSoup("<div></div>", "html.parser")
    wrapper = wrapper_soup.div

    # Gộp node vào wrapper
    if isinstance(nodes, list):
        for n in nodes:
            if isinstance(n, Tag):
                wrapper.append(n)
    elif isinstance(nodes, Tag):
        wrapper.append(nodes)
    else:
        frag = BeautifulSoup(str(nodes), "html.parser")
        for n in list(frag.contents):
            wrapper.append(n)
    # Chỉ giữ Question + Answer + Explanation
    # Xóa toàn bộ node phía sau Answer
    answers = wrapper.select(".collapseomatic_content")
    if answers:
        last_answer = answers[-1]
        for sib in list(last_answer.next_siblings):
            sib.extract()
    # Xóa quảng cáo
    for bad in wrapper.select(
        ".sf-mobile-ads, style, iframe, svg, ins, "
        ".adsbygoogle, [id^='sf-ads'], [class*='av-pos']"
    ):
        bad.decompose()

    # Xóa script (trừ math/tex)
    for s in wrapper.find_all("script"):
        if s.get("type") != "math/tex":
            s.decompose()
    # Xóa block chỉ chứa chữ advertisement
    for tag in wrapper.find_all(True):
        if tag.get_text(" ", strip=True).lower() == "advertisement":
            tag.decompose()
    # Bỏ wrapper Google annotation
    for a in wrapper.select("a.google-anno"):
        text_span = a.select_one(".google-anno-t")
        replacement = text_span.get_text() if text_span else a.get_text()
        a.replace_with(replacement)

    for span in wrapper.select(".google-anno-t"):
        span.replace_with(span.get_text())
    # Xóa nút View Answer
    for tog in wrapper.select(".collapseomatic"):
        tog.decompose()
    # Chỉ giữ các attribute cần thiết
    for tag in wrapper.find_all(True):
        keep = {}
        if tag.name == "a":
            href = tag.get("href")
            if href and href != "#":
                keep["href"] = href
        elif tag.name == "script":
            t = tag.get("type")
            if t:
                keep["type"] = t
        elif tag.name == "math":
            d = tag.get("display")
            if d:
                keep["display"] = d

        tag.attrs = keep

    # Xóa tag rỗng (từ trong ra ngoài)
    while True:
        removed = False

        for tag in reversed(wrapper.find_all(True)):
            if tag.name in ("br", "img"):
                continue
            if not tag.get_text(strip=True) and not tag.find(True):
                tag.decompose()
                removed = True
        if not removed:
            break
    return wrapper
```

```python
import re
from bs4 import NavigableString, Tag

_MATH_UNICODE_MAP = {
    "√": r"\sqrt",
    "±": r"\pm",
    "∓": r"\mp",
    "×": r"\times",
    "÷": r"\div",
    "≤": r"\leq",
    "≥": r"\geq",
    "≠": r"\neq",
    "≈": r"\approx",
    "≡": r"\equiv",
    "∞": r"\infty",
    "∑": r"\sum",
    "∏": r"\prod",
    "∫": r"\int",
    "∂": r"\partial",
    "∇": r"\nabla",
    "∈": r"\in",
    "∉": r"\notin",
    "⊂": r"\subset",
    "⊆": r"\subseteq",
    "∪": r"\cup",
    "∩": r"\cap",
    "→": r"\to",
    "⇒": r"\Rightarrow",
    "⇔": r"\Leftrightarrow",
    "α": r"\alpha",
    "β": r"\beta",
    "γ": r"\gamma",
    "δ": r"\delta",
    "θ": r"\theta",
    "λ": r"\lambda",
    "μ": r"\mu",
    "π": r"\pi",
    "σ": r"\sigma",
    "φ": r"\phi",
    "ω": r"\omega",
    "Δ": r"\Delta",
}

def convert_math(text: str) -> str:
    """
    Chuẩn hóa biểu thức toán học trong text.
    """
    # \( ... \) -> $...$
    text = re.sub(
        r"\\\((.+?)\\\)",
        lambda m: f"${m.group(1).strip()}$",
        text,
        flags=re.S,
    )
    # \[ ... \] -> $$...$$
    text = re.sub(
        r"\\\[(.+?)\\\]",
        lambda m: f"$${m.group(1).strip()}$$",
        text,
        flags=re.S,
    )
    # °
    text = re.sub(r"(\w)°", r"\1$^{\\circ}$", text)
    text = text.replace("°", "$^{\\circ}$")

    # Unicode math
    for symbol, latex in _MATH_UNICODE_MAP.items():
        text = text.replace(symbol, f"${latex}$")
    return text

_WORD_END_RE = re.compile(r"([A-Za-z0-9\)\]\}])$")


def html2markdown(node: Tag) -> str:
    """
    Convert BeautifulSoup Tag -> Markdown.
    """
    def render(n):
        if isinstance(n, NavigableString):
            return convert_math(str(n))
        if not isinstance(n, Tag):
            return ""
        name = n.name.lower()
        if name == "br":
            return "\n"
        if name == "math":
            expr = n.get_text(strip=True)
            return (
                f"$${expr}$$"
                if n.get("display") == "block"
                else f"${expr}$"
            )
        if name == "script":
            t = n.get("type", "")
            if "math/tex" in t:
                expr = n.get_text(strip=True)
                return (
                    f"$${expr}$$"
                    if "mode=display" in t
                    else f"${expr}$"
                )
            return ""
        parts = []
        for child in n.children:
            if isinstance(child, Tag) and child.name.lower() in {"sup", "sub"}:
                symbol = "^" if child.name.lower() == "sup" else "_"
                script = render(child)
                if parts:
                    last = parts[-1]
                    m = _WORD_END_RE.search(last)
                    if m:
                        prefix = last[:m.start()]
                        base = last[m.start():]
                        if base.startswith("$") and base.endswith("$"):
                            base = base[1:-1]
                        parts[-1] = (
                            f"{prefix}${base}{symbol}{{{script}}}$"
                        )
                        continue
                parts.append(f"${symbol}{{{script}}}$")
            else:
                parts.append(render(child))

        inner = "".join(parts)
        if name in {"b", "strong"}:
            return f"**{inner}**"
        if name in {"i", "em"}:
            return f"*{inner}*"
        if name == "code":
            return f"`{inner}`"
        if name == "a":
            href = n.get("href")
            return f"[{inner}]({href})" if href else inner
        return inner

    return render(node).strip()
```
